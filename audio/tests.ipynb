```python
from pydub import AudioSegment
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot signal from webm file
sound = AudioSegment.from_file("../../data/audio/soft_10s/Arctic Monkeys - I Wanna Be Yours.webm", format="webm")
```

```python
channels = sound.channels
```

```python
# Plot signal
plt.figure(figsize=(20, 5))
plt.plot(sound.get_array_of_samples())
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Apply fft to signal
arr = sound.get_array_of_samples()
# Slice signal to 1/4th
arr = arr[:int(len(arr)/4)]
x = np.fft.fft(arr)
x = np.abs(x)
# Plot frequency vs amplitude
plt.figure(figsize=(20, 5))
plt.plot(x)
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Get dominant frequency
max_freq = np.argmax(x)
print("Dominant frequency: ", max_freq)
```

```python
# Only keep dominant frequency
x[max_freq+1:] = 0
x[:max_freq-1] = 0
# Plot frequency vs amplitude
plt.figure(figsize=(20, 5))
plt.plot(x)
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
