```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import torch
import numpy as np
from torch.utils.data import TensorDataset, DataLoader
from read_data import get_compressed_object, getImageFromString
from torchvision import transforms
import io
from PIL import Image
```

```python
def make_dataset(root):
    INCH_TO_CM = 2.54
    LBS_TO_G = 453.592
    files = os.listdir(root)
    f = os.path.join(root,files[0])
    record = get_compressed_object(f)
    img_data = cv2.imdecode(np.frombuffer(record['image_data'], np.uint8), -1)
    img_data = img_data.reshape(1,img_data.shape[0],img_data.shape[1],img_data.shape[2])
    dims = record['dimensions']
    dimens_data = np.array(dims)
    dimens_data = dimens_data.astype('float64')
#     print(dimens_data)
    dimens_data = INCH_TO_CM*dimens_data
    weight = float(record['weight'])
#     print(type(weight))
    weight = weight*LBS_TO_G
    dimens_data = (np.append(dimens_data,weight)).reshape(1,4)
    for i in range(1,len(files)):
        f = os.path.join(root,files[i])
        record = get_compressed_object(f)
        im = cv2.imdecode(np.frombuffer(record['image_data'], np.uint8), -1)
        im = im.reshape(1,im.shape[0],im.shape[1],im.shape[2])
        img_data = np.append(img_data,im,axis=0)
        # print(im)
        # print(im.shape)
        # thickness_mask = record['thickness_mask']
        # sp = plt.subplot(121)
        # plt.imshow(im)
        # sp.set_xlabel(str(record['weight'])+' lbs')
        # sp=plt.subplot(122)
        # plt.imshow(thickness_mask)
        dims = record['dimensions']
        temp_dimens_data = np.array(dims)
        temp_dimens_data = temp_dimens_data.astype('float64')
        temp_dimens_data = INCH_TO_CM*temp_dimens_data
        temp_weight = float(record['weight'])
        temp_weight = temp_weight*LBS_TO_G
        temp_dimens_data = (np.append(temp_dimens_data,weight)).reshape(1,4)
        dimens_data = np.append(dimens_data, temp_dimens_data, axis=0)
        # dimstr = ' inches by '.join(dims)+' inches'
        # sp.set_xlabel(dimstr)
        # matplotlib.use('TkAgg')
#         plt.show()
    tensor_img = torch.Tensor(img_data) # transform to torch tensor
    tensor_dimens = torch.Tensor(dimens_data)
    
    dataset = TensorDataset(tensor_img,tensor_dimens) # create your dataset
    return dataset
```
