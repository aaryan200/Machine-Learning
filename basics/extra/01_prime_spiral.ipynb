```python
import numpy as np
import matplotlib.pyplot as plt
```

Sieve for finding all prime numbers up to a given limit

```python
def sieveOfErasto(N: int):
    primes = np.ones(N, dtype=bool)
    primes[0] = False
    primes[1] = False
    i = 2
    while i*i <= N:
        if primes[i]:
            for j in range(i*i, N, i):
                primes[j] = False
        i += 1
    # Return the list of primes
    return np.where(primes)[0]

sieveOfErasto(100)
```

### Plot the prime numbers in $(r, \theta)$ format

```python
def plot_prime_rth(N: int):
    primes = sieveOfErasto(N)
    print(f"{len(primes)} primes found")
    fig, ax = plt.subplots(figsize=(10, 6))
    plt.scatter(primes*np.cos(primes), primes*np.sin(primes), color='#068f94', s = 4)
    plt.xlim(-N+10, N+10)
    plt.ylim(-N+10, N+10)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```
