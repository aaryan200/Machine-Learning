```python
def j_th(th,x,y):
    import numpy as np
    m=x.shape[0]
    n=x.shape[1]-1
    temp=1.0/(1+np.exp(-1*(x@th)))
    out=(y*(np.log(temp))+(1-y)*(np.log(1-temp)))
    s=float(np.sum(out))
    return(-1*s/m)
def newt_method(x,y):
    EPS=1.0e-10
    m=x.shape[0]
    n=x.shape[1]-1
    th=np.zeros(n+1)
    th=th.reshape(n+1,1)
    conv=np.Infinity
    j0=j_th(th,x,y)
    it=0
    while abs(conv)>EPS:
        it+=1
        temp1=1.0/(1.0+np.exp(-1*x@th))
        H=((np.transpose(x))@((temp1)*(1-temp1)*x))
        H=H/m
        dJ=np.sum(((temp1-y)*x/m),axis=0)
        dJ=dJ.reshape(n+1,1)
        temp2=(np.linalg.pinv(H))@dJ
#         print(np.linalg.pinv(H))
#         print(dJ)
#         print(temp2)
        th=th-temp2
        j1=j_th(th,x,y)
        conv=j1-j0
        j0=j1
    print(it)
    return th
```
