```python
def dot_val(th,X):
    """
    th - (n+1)x1
    X - 1x(n+1)
    """ 
    import numpy as np
    out=X@th
    return out[0][0]
def hval(th,X):
    """
    th - (n+1)x1
    X - 1x(n+1)
    """ 
    from math import exp
    out=1/(1+exp((-1)*dot_val(th,X)))
    return out
def probab(th,X,y):
    """
    X - mx(n+1)
    y- mx1
    th - (n+1)x1
    """
    from math import log
    ZERO=1.0e-16
    m=y.shape[0]
    n=th.shape[0]-1
    s=0.0
    for i in range(m):
        t=X[i].reshape(1,n+1)
        ht=hval(th,t)
        temp=y[i][0]
#         if (abs(ht)<=ZERO or abs(1-ht)<=ZERO):
#             return 'c'
        #print(ht)
        s+=temp*log(ht)+(1-temp)*log(1-ht)
    return s
def pred(X,y):
    """
    X - mx(n+1)
    y- mx1
    test - 1x(n+1)
    """
    ERR=1.0e-3
    import numpy as np
    alp=1.0e-4
    m=y.shape[0]
    n=X.shape[1]-1
    th=np.zeros(n+1)
    th=th.reshape(n+1,1)
    th0=np.zeros(n+1)
    th0=th0.reshape(n+1,1)
    j0=probab(th,X,y)
    conv=j0
    print(conv)
    it=0
    while (abs(conv)>=ERR):
        for j in range(n+1):
            t=0.0
            for i in range(m):
                t1=X[i].reshape(1,n+1)
                t+=(y[i]-hval(th0,t1))*X[i][j]
            th[j]=th0[j]+alp*t
        j1=probab(th,X,y)
        conv=j1-j0
        print(conv)
        j0=j1
        th0=th
        it+=1
    return th
```

```python
def min_max_test(x,y,th):
    """
    th - (n+1)x1
    x - mx(n+1)
    y - mx1
    """
    m=y.shape[0]
    n=x.shape[1]-1
    import numpy as np
    max_1=(-1)*np.Infinity
    min_1=np.Infinity
    for i in range(m):
        temp=x[i].reshape(1,n+1)
        hv=hval(th,temp)
        if (y[i]==1 and hv<min_1):
            min_1=hv
        elif (y[i]==0 and hv>max_1):
            max_1=hv
    return(min_1,max_1)
```
