```python
%%timeit -n 100
import numpy as np
def add_col(A,m,i,j):
    n=A.shape[1]
    E=np.eye(n)
    if i==j:
        E[i,i]=m+1
    else:
        E[j,i]=m
    return A@E
def det1(A):
    n=A.shape[0]
    for i in range(n):
        for j in range(n):
            A[i,j]=float(A[i,j])
    d=1
    for i in range(n):
        if abs(A[i,i])<=1.0e-16:
            x=0
            for col in range(i+1,n):
                if A[i,col]!=0:
                    A=add_col(A,1,i,col)
                    break
                else:
                    x+=1
            if x==n-1-i:
                d=0
                break
            else:
                for row in range(i+1,n):
                    for col in range(n-1,i-1,-1):
                        A[row,col]-=A[row,i]*A[i,col]/A[i,i]
                d=d*A[i,i]
        else:
            for row in range(i+1,n):
                for col in range(n-1,i-1,-1):
                    A[row,col]-=A[row,i]*A[i,col]/A[i,i]
            d=d*A[i,i]
    d=round(d,5)
    return d
def inv(A):
    n=A.shape[0]
    d=det1(A)
    if d!=0:
        A_inv=np.array([])
        for k in range(n**2):
            A_inv=np.append(A_inv,0.0)
        A_inv=A_inv.reshape(n,n)
        for i in range(n):
            for j in range(n):
                B=np.array([])
                for k in range((n-1)**2):
                    B=np.append(B,0.0)
                B=B.reshape(n-1,n-1)
                for row in range(i):
                    for col in range(j):
                        B[row,col]=A[row,col]
                    for col in range(j+1,n):
                        B[row,col-1]=A[row,col]
                for row in range(i+1,n):
                    for col in range(0,j):
                        B[row-1,col]=A[row,col]
                    for col in range(j+1,n):
                        B[row-1,col-1]=A[row,col]
                A_inv[i,j]=((-1)**(i+j))*det1(B)/d
                A_inv[i,j]=round(A_inv[i,j],5)
        A_inv=A_inv. T
        return A_inv
    if abs(d)<=1.0e-16:
        s="Inverse is NOT possible:This is a singular matrix"
        return s
F=np.array([[1.0,2.0,3.0],[9.0,29.0,19.0],[200.0,39.0,18.0]])
G=inv(F)
```

# Conversion of decimal number to 32 bit single precision IEEE-754 floating point representation

```python
def list_inv(a):
    n=len(a)
    b=[]
    for i in range(n):
        b.append(a[n-1-i])
    return b
a=[1,2,3]
print(list_inv(a))
```

```python
def int_bin(n):
    li=[]
    while n>0:
        a=n%2
        li.append(a)
        n=int(n/2)
    li=list_inv(li)
    return li
l=int_bin(25)
print(l)
```

```python
def fr_bin(f):
    i=0
    li=[]
    while (f>0 and i<35):
        f=2*f
        if f>=1:
            li.append(1)
            f=f-1
            i+=1
        elif f<1:
            li.append(0)
            i+=1
    return li
print(fr_bin(0.374))
```

```python
def sign_bit(n):
    li=[]
    if n>=0:
        li.append(0)
    if n<0:
        li.append(1)
    return li
```

```python
print(int_bin(1))
print(fr_bin(0.0))
```

```python
def exp_bit(n):
    n=abs(n)
    if n<1:
        a=fr_bin(n)
        i=0
        while a[i]==0:
            i+=1
        i+=1
        bias_exp=127-i
        bin_bias_exp=int_bin(bias_exp)
        j=0
        while len(bin_bias_exp)<8:
            bin_bias_exp.insert(j,0)
            j+=1
        return bin_bias_exp
    elif n>=1:
        int_n=int(n)
        b=int_bin(int_n)
        x=len(b)
        bia_exp=126+x
        bin_bia_exp=int_bin(bia_exp)
        k=0
        while len(bin_bia_exp)<8:
            bin_bia_exp.insert(k,0)
            k+=1
        return bin_bia_exp
print(exp_bit(0.5))
```

```python
def mantissa_bits(n):
    n=abs(n)
    if n<1:
        a=fr_bin(n)
        i=0
        while a[i]==0:
            del a[i]
            i+=1
        del a[i]
        while len(a)>23:
            del a[len(a)-1]
        return a
    elif n>=1:
        b=int(n)
        fr=n-b
        li=int_bin(b)+fr_bin(fr)
        del li[0]
        while len(li)>23:
            del li[len(li)-1]
        return li
```

```python
n=.08429328
print(sign_bit(n))
print(exp_bit(n))
print(mantissa_bits(n))
print('\n')
a=-1.008429328
print(sign_bit(a))
print(exp_bit(a))
print(mantissa_bits(a))
```
