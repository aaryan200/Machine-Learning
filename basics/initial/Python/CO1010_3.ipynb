```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.interpolate as si
x = np.arange(1,20,2)
y = np.exp(x/2.0)
f = si.interp1d(x,y,'cubic')
print(f)
xnew = np.arange(1, 19,0.1)
ynew = f(xnew)
plt.plot(x, y, 'o', xnew, ynew, '-')
```

```python
# f(x) x^2 -10<=x<=10
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y,'k-')
```

```python
# f1(x) = x**2 and f2(x)=exp(-x**2)
y1 = x**2
y2 = np.exp(-x**2)
plt.plot(x,y1,'ko-')
plt.plot(x,y2,'b*-')
```

```python
plt.ylim(0,1.5)
plt.plot(x,y1,'ko-',label='quadratic')
plt.plot(x,y2,'b*-',label='Gaussian')
plt.grid()
plt.xlabel('x')
plt.ylabel('y1,y2')
plt.legend()
```
