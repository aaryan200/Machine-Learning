```python
def way_li(n):
    li=[]
    if n==1:
        li1=[1]
        li.append(li1)
        return li
    elif n>=2:
        x=int(n/2)
        for i in range(1,x+1):
            l=way_li(n-i)[:]
            for j in l:
                if i<=min(j):
                    j.insert(0,i)
                    li.append(j)
        a=[n]
        li.append(a)
        return li
def print_way(n):
    li=way_li(n)
    for j in li:
        print(*j,sep='+')
    return
```

```python
way_li(5)
```

```python
li=[5,6]
print(*li,sep='+')
```

```python
print_way(21)
```
