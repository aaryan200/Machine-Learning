```python
# Dropping k_value, l_value, m_value, n_value features of training data
x2=x1.drop(['k_value','l_value','m_value','n_value'],axis=1)

# Extracting quality of training data in numpy array "y"
y1=x2['quality']
y2=y1.to_numpy()
y=y2.reshape(1280,1)

# Removing quality from training data
x3=x2.drop('quality',axis=1).to_numpy() 

# Test data from file
te1=al1[960:1279]    

# Extracting quality of test data in numpy array "teans"
teans1=te1['quality']   
teans2=teans1.to_numpy()
teans=teans2.reshape(319,1)

# Since 'quality','k_value','l_value','m_value','n_value' are removed from training data, removing it from
# test data
te2=te1.drop(['quality','k_value','l_value','m_value','n_value'],axis=1).to_numpy()

# Concatenating 1.0 at start of every training data ...to apply logistic regression..
x=np.c_[np.ones(1280),x3]
te=np.c_[np.ones(319),te2]   # Final test data
```

```python
# Removing outliers from training data
temp1=np.percentile(x,1,axis=0)
temp99=np.percentile(x,99,axis=0)
c1=0
for col in range(13):
    p_1=temp1[col]
    p_99=temp99[col]
    for i in range(1280):
        if x[i][col]>p_99:
            x=np.delete(x,i,axis=0)
            y=np.delete(y,i,axis=0)
            c1+=1
        elif x[i][col]<p_1:
            x=np.delete(x,i,axis=0)
            y=np.delete(y,i,axis=0)
            c1+=1
        if (i==1279-c1):
            break
```

```python
# Normalization of training data along with test data
n_row=x.shape[0]
for col in range(1,13):
    min_col=np.Infinity
    max_col=(-1)*np.Infinity
    for row in range(n_row):
        temp1=x[row][col]
        if (temp1>max_col):
            max_col=temp1
        elif (temp1<min_col):
            min_col=temp1
    x[:,col]=(x[:,col]-min_col)/(max_col-min_col)
    te[:,col]=(te[:,col]-min_col)/(max_col-min_col)
```

```python
def j_th(Î¸,x,y):
    """
    return the negative of log likelihood function
    =============================================
    return type - float64
    """
    import numpy as np
    m=x.shape[0]    # Number of training data
    n=x.shape[1]-1  # Number of features
    temp=1.0/(1+np.exp(-1*(x@Î¸)))
    out=(y*(np.log(temp))+(1-y)*(np.log(1-temp)))
    s=float(np.sum(out))
    return(-1*s/m)


def newton_method(x,y):
    """
    return the array of required parameters for classification problem
    ==================================================================
    """
    EPS=1.0e-12    #Convergence limit 
    m=x.shape[0]
    n=x.shape[1]-1
    Î¸=np.zeros(n+1)
    Î¸=Î¸.reshape(n+1,1)
    conv=np.Infinity
    j0=j_th(Î¸,x,y)
    it=0
    while abs(conv)>EPS:
        it+=1
        temp1=1.0/(1.0+np.exp(-1*x@Î¸))
        H=((np.transpose(x))@((temp1)*(1-temp1)*x))
        H=H/m
        dJ=np.sum(((temp1-y)*x/m),axis=0)
        dJ=dJ.reshape(n+1,1)
        temp2=(np.linalg.pinv(H))@dJ
        Î¸=Î¸-temp2
        j1=j_th(Î¸,x,y)
        conv=j1-j0
        j0=j1
    return Î¸
```

```python
Î¸1=newton_method(x,y)
```

```python
#Checking accuracy of model
count=0
my_ans=1.0/(1.0+np.exp(-1*te@Î¸1))

for i in range(319):
    if (my_ans[i]>=0.5 and teans[i]==1.0):
        count+=1
    elif (my_ans[i]<0.5 and teans[i]==0.0):
        count+=1
print(f'Accuracy is {count*100/319}%')
```
