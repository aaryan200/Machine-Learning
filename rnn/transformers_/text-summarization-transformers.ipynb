```python
from transformers import pipeline
```

```python
summarizer = pipeline('summarization')
```

```python
article = """
Memory Efficiency:

Iterators provide a memory-efficient way to iterate over elements, especially for large datasets or infinite sequences.
They generate elements on-the-fly and don't require storing the entire sequence in memory.
Generalization:

Not all iterable objects support direct indexing. For example, sets and dictionaries in Python are iterable but don't have a defined order, so you can't access elements by index.
Iterators provide a consistent way to iterate over various types of iterable objects.
Custom Iteration Logic:

Iterators allow you to define custom iteration logic. You can control how elements are generated or processed during iteration.
This is useful for lazy evaluation or for cases where the iteration logic is more complex than simple sequential access.
Support for Infinite Sequences:

Iterators can be used to represent and iterate over infinite sequences, something that is not feasible with direct indexing.
Compatibility:

Some APIs or libraries may provide iterators rather than sequences with direct indexing. Understanding iterators allows you to work with a broader range of Python tools..
"""
```

```python
# max_length and min_length denotes the range of length of output
# If do_sample is True, it will involve something random in output and if False, it will do
# greedy sampling
summarizer(article, max_length = 130, min_length = 30, do_sample = False)
```
